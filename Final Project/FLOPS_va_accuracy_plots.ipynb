### ***FLOPs vs accuracy visualization per model***

This code visualizes the relationship between the test accuracy and the total fine-tuning FLOPs (Floating Point Operations) of various models. The scatter plot highlights the performance of each model in terms of accuracy and computational cost during fine-tuning. <br> Models include EfficientNet, ResNet152x4, and a couple of variations of Vision Transformers (ViT_B_16, ViT_B_32), with pre-trained configurations.<br> A logarithmic scale is applied to the FLOPs axis to better illustrate the differences across models with varying computational complexities. Labels and colors are added for clarity and to facilitate comparisons between models.

```python
import matplotlib.pyplot as plt

# Define model data (Test Accuracy and Total Fine-Tuning FLOPs)
models = [
    {"name": "EffNet", "accuracy": 8.25, "flops": 1.67e+17},  # Example fine-tuning FLOPs
    {"name": "ResNet152x4", "accuracy": 86.00, "flops": 5.87e+18},  # Example fine-tuning FLOPs
    {"name": "ViT-B_32", "accuracy": 82.75, "flops": 2.79e+17},  # Example fine-tuning FLOPs
    {"name": "ViT-B_16", "accuracy": 85.25, "flops": 1.08e+18},  # Example fine-tuning FLOPs
    {"name": "R50 ViT-B_16", "accuracy": 87.75, "flops": 1.34e+18},  # Example fine-tuning FLOPs
]

# Extract values for plotting
names = [model["name"] for model in models]
accuracies = [model["accuracy"] for model in models]
flops = [model["flops"] for model in models]
colors = ['blue', 'green', 'orange', 'red', 'purple']

# Create scatter plot
plt.figure(figsize=(10, 6))
for i, model in enumerate(models):
    plt.scatter(flops[i], accuracies[i], color=colors[i], s=100, label=model["name"], alpha=0.7)
    # Place labels differently for ResNet152x4 and ViT-B_16
    if model["name"] in ["ResNet152x4", "ViT-B_16"]:
        plt.text(flops[i] * 0.9, accuracies[i], model["name"], fontsize=9, ha='right', color='black')
    else:
        plt.text(flops[i] * 1.1, accuracies[i], model["name"], fontsize=9, ha='left', color='black')

# Set plot details
plt.title("Accuracy vs Total Fine-Tuning FLOPs for Different Models")
plt.xlabel("Total Fine-Tuning FLOPs (log scale)")
plt.ylabel("Test Accuracy (%)")
plt.xscale("log")  # Use log scale for FLOPs
plt.grid(True, linestyle='--', alpha=0.6)

# Add legend for colors
plt.legend(title="Models", loc='lower right')

# Show plot
plt.tight_layout()
plt.show()
```
